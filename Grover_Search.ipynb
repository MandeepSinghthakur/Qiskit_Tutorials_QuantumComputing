# Grover Search for boolean satisfiability

Imagine you are inviting some friends to a party, some 
who are coulpes, and some who are not on speaking terms. Specifically, Alice and Bob are in relationship, as are Carol and David. However, Alice and David had a breakup a while ago and haven't been civil with each other since. Armed with quantum computer and Qiskit Aqua, how can you leverage Grover search to identify combination of people to invire ?

And Grover Search helps you with it - It can be described as Boolean Satisfiability problem
((A and B) or (C and D)) and not (A and D)


```python
from qiskit import BasicAer
from qiskit.aqua.algorithms import Grover
from qiskit.aqua.components.oracles import LogicalExpressionOracle
from qiskit.tools.visualization import plot_histogram
```

```python
log_expr = '((A & B) | (C & D)) & ~(A & D)'

algorithm = Grover(LogicalExpressionOracle(log_expr))
```

Now we' run the algorithm on a simulator, printing the result that occured most often. This result is expressed as the numeric representations of our four friends; a minus sign indicationg which ones grover advised against inviting in that particular result

```python
# Run the algorithm on a simulator, printing the most 
# frequently occuring result
backend = BasicAer.get_backend('qasm_simulator')
result = algorithm.run(backend)
print(result['top_measurement'])
```

```python
plot_histogram(result['measurement'])
```
